If my goal is to retire before age 46, which school offers a better return on investment: CU Boulder or Colorado School of Mines?

Assumptions:
This simulation models 30 hypothetical students from each school (CU Boulder and Colorado School of Mines) using randomly generated tuition, housing, and salary values based on realistic 2024–2025 averages. Each student is assumed to complete a four-year bachelor’s degree with stable yearly tuition and living costs, which include housing, food, and utilities. All graduates are assumed to secure employment immediately after graduation, and the analysis focuses on their starting salaries without accounting for future raises, inflation, taxes, or student loan interest. Return on investment (ROI) and years to recoup educational costs are calculated from each student’s simulated expenses and starting salary, providing an approximate comparison of financial outcomes between the two schools.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# parameters
n_students = 50  # updated from 30 to 50
years = 4

# CU Boulder simulated data
cu_boulder = pd.DataFrame({
    "School": ["CU Boulder"] * n_students,
    "Tuition_per_year": np.random.normal(16000, 1000, n_students).round(),
    "Housing_per_year": np.random.normal(14000, 1500, n_students).round(),
    "Average_starting_salary": np.random.normal(72000, 7000, n_students).round()
})

# Colorado School of Mines simulated data
mines = pd.DataFrame({
    "School": ["Colorado School of Mines"] * n_students,
    "Tuition_per_year": np.random.normal(19000, 1000, n_students).round(),
    "Housing_per_year": np.random.normal(15000, 1200, n_students).round(),
    "Average_starting_salary": np.random.normal(85000, 8000, n_students).round()
})

# Combine
df = pd.concat([cu_boulder, mines], ignore_index=True)

# Calculate costs and ROI
df["Total_cost"] = (df["Tuition_per_year"] + df["Housing_per_year"]) * years
df["ROI_ratio"] = df["Average_starting_salary"] / (df["Tuition_per_year"] + df["Housing_per_year"])
df["Years_to_recoup"] = df["Total_cost"] / df["Average_starting_salary"]

# --- Summary statistics ---
summary = df.groupby("School")[["Tuition_per_year", "Housing_per_year", "Average_starting_salary", "ROI_ratio", "Years_to_recoup"]].mean().round(2)
print("\n--- Average Metrics per School (50 simulated students each) ---")
print(summary)

# --- Plot: Distribution of Starting Salaries ---
plt.figure(figsize=(10,6))
for school, color in zip(df["School"].unique(), ["#1f77b4", "#ff7f0e"]):
    subset = df[df["School"] == school]
    plt.hist(subset["Average_starting_salary"], bins=10, alpha=0.6, label=school, color=color)
plt.title("Distribution of Starting Salaries (Simulated Students)")
plt.xlabel("Starting Salary (USD)")
plt.ylabel("Number of Students")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()

# --- Plot: ROI Comparison ---
plt.figure(figsize=(10,6))
df.boxplot(column="ROI_ratio", by="School", grid=False, patch_artist=True,
           boxprops=dict(facecolor="#c6dbef", color="#3182bd"),
           medianprops=dict(color="red", linewidth=2))
plt.title("ROI Ratio Comparison Between Schools")
plt.suptitle("")
plt.ylabel("ROI (Salary / Annual Cost)")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
