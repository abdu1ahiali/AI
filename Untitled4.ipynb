Is the average view duration of YouTube videos is more than 4 minutes.

Null hypothesis 

: Average AVD ≤ 4 min

Alternative hypothesis 

: Average AVD > 4 min

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



pop_mean = 3.5        # minutes
pop_sd = 2.0          # standard deviation of AVD
np.random.seed(42)

population = np.random.normal(pop_mean, pop_sd, 500000)
population = np.clip(population, 0, None)  # Remove negative watch times


sample_size = 100
n_samples = 1000

sample_means = []

for _ in range(n_samples):
    sample = np.random.choice(population, sample_size, replace=True)
    sample_means.append(sample.mean())

sample_means = np.array(sample_means)


df = pd.DataFrame({"sample_mean": sample_means})

SE = df["sample_mean"].std(ddof=1)
MOE_95 = 1.96 * SE

print("Estimated mean of means:", df['sample_mean'].mean())
print("Standard error:", SE)
print("95% Margin of Error (±):", MOE_95)

plt.figure(figsize=(10,5))
plt.hist(df["sample_mean"], bins=40, color='skyblue', edgecolor='black')
plt.axvline(df["sample_mean"].mean(), color='red', linewidth=2, label="Mean of sample means")
plt.title("Sampling Distribution of Sample Means (n=100, 1000 samples)")
plt.xlabel("Sample Mean (minutes)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
