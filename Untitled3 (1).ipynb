What proportion of Wikipedia biographies include more than 3 categories?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simulated sample of 30 biographies
data = {
    "title": [
        "Ada_Lovelace", "Albert_Einstein", "Marie_Curie", "Isaac_Newton", "Leonardo_da_Vinci",
        "Nikola_Tesla", "Charles_Darwin", "Galileo_Galilei", "Stephen_Hawking", "Rosalind_Franklin",
        "Alan_Turing", "Cleopatra", "Mahatma_Gandhi", "Martin_Luther_King_Jr.", "Winston_Churchill",
        "Frida_Kahlo", "Ludwig_van_Beethoven", "Pablo_Picasso", "Nelson_Mandela", "Joan_of_Arc",
        "Genghis_Khan", "Queen_Victoria", "Abraham_Lincoln", "Emmeline_Pankhurst", "Socrates",
        "Confucius", "Florence_Nightingale", "Alexander_the_Great", "Malala_Yousafzai", "Barack_Obama"
    ],
    "num_categories": [
        5,8,7,6,4,5,6,4,5,3,6,2,5,4,6,3,5,3,6,2,4,5,4,3,2,2,3,5,4,6
    ]
}

# Create DataFrame
df = pd.DataFrame(data)
df["more_than_3_categories"] = df["num_categories"] > 3

# Calculate proportion and 95% CI
n = len(df)
prop = df["more_than_3_categories"].mean()
se = np.sqrt(prop*(1-prop)/n)
ci_low = prop - 1.96*se
ci_high = prop + 1.96*se

print(df)
print("\nSample size:", n)
print("Proportion with >3 categories:", round(prop,3))
print("95% CI:", (round(ci_low,3), round(ci_high,3)))

# Bar graph
counts = df["more_than_3_categories"].value_counts()
bars = plt.bar(["â‰¤3 categories", ">3 categories"],
        [counts.get(False,0), counts.get(True,0)],
        color=["red","green"])

plt.title("Wikipedia Biographies: Number of Categories (>3)")
plt.ylabel("Number of Articles")

# Annotate graph with proportion, CI, and sample size
plt.text(0.5, max(counts)+1, f"Mean proportion (>3 categories): {prop:.3f}", ha='center', fontsize=10)
plt.text(0.5, max(counts)+0.5, f"95% CI: [{ci_low:.3f}, {ci_high:.3f}]", ha='center', fontsize=10)
plt.text(0.5, max(counts)+0, f"Sample size: {n}", ha='center', fontsize=10)

plt.ylim(0, max(counts)+5)
plt.show()
```
