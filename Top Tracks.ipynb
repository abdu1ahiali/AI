```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = {
    'gender': ['Male', 'Female', 'Female', 'Male', 'Other', 'Female']
}
df = pd.DataFrame(data)

# Create frequency table
frequency_table = df['gender'].value_counts()

# Print frequency table
print(frequency_table)

# Plot using matplotlib
frequency_table.plot(kind='bar', color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Frequency of Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("Spotifydata.csv")

# Check column names to find the genre column
print("Columns in dataset:", df.columns)

# Replace 'genre' with your actual genre column name if different
genre_col = 'genre'  # Change this if needed

# Create frequency table
freq_table = df[genre_col].value_counts()

# Create bar graph
freq_table.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frequency of Music Genres in Spotify Editorial Playlists')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Answer the question: Which genre appears most frequently?
most_frequent_genre = freq_table.idxmax()
most_frequent_count = freq_table.max()

print(f"\nAnswer: The most frequent genre in Spotify's editorial playlists is '{most_frequent_genre}' with {most_frequent_count} songs.")
```
